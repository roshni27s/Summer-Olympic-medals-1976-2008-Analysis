```python


import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
games=pd.read_csv(r'C:\Users\roshn\data_visualiasation_and_ML\Summer-Olympic-medals-1976-to-2008.csv',encoding='Latin 1')
```

```python
games.head()
```

```python
# checking null values
games.isnull().sum()
```

```python
# dropping null values
games.dropna(inplace=True)
```

```python
games.info()
```

```python
# converting float values in year to int
games['Year'] = games['Year'].astype(int)
```

```python
games.head()
```

```python
# organising the name of the athelete as firstname and lastname
games.Athlete=games.Athlete.str.split(', ').map(lambda x : ' '.join(x[::-1]))
```

```python
games.head()
```

```python
# making a new column for medal ranks
games['Medal_rank']=games['Medal']
```

```python
rank={'Gold':1,'Silver':2,'Bronze':3}
games['Medal_rank']=games['Medal_rank'].map(rank)
```

```python
games.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('dark_background')
```

```python
sns.countplot(y='City',data=games)
```

* ### Beijing has the maximum number of participants and Montreal has the lowest.

```python
# yearwise data
games.groupby('Year').count()
```

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=90)
sns.countplot(x='Sport',data=games)
```

* ### Aquatics sport had maximum number of participants

```python
plt.figure(figsize=(8,8))
sns.countplot(y='City',hue='Medal',data=games,palette='coolwarm')
```

* ### Beijing has max number of gold medals
* ### Sydney has max number of silver medals
* ### Beijing has max number of bronze medals

```python
plt.figure(figsize=(8,8))
sns.countplot(y='City',hue='Gender',data=games,palette='Set1')
```

* ### Beijing had max female participation
* ### Sydney had max male participation
